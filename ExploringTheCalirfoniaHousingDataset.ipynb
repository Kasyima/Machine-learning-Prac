# `Linear regression`

```python
housing_data = housing_data.dropna()
```

```python
housing_data.loc[housing_data['median_house_value'] ==  500001].count()
```

```python
housing_data = housing_data.drop(housing_data.loc[housing_data['median_house_value'] == 500001].index)
```

```python
housing_data['ocean_proximity'].unique()
```

* Use one hot encoding to convert categorical data to numeric

```python
housing_data = pd.get_dummies(housing_data, columns=['ocean_proximity'])
```

```python
X = housing_data.drop('median_house_value', axis=1)
Y = housing_data['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lnrreg = LinearRegression(normalize=True)

clsf = lnrreg.fit(x_train, y_train)
```

```python
score = clsf.score(x_train, y_train)
score
```

```python
predictors = X.columns
predictors
```

```python
coefficients = pd.Series(clsf.coef_,predictors).sort_values()
coefficients
```

```python
predicted = clsf.predict(x_test)
```

```python
from sklearn.metrics import r2_score
scoress = r2_score(y_test, predicted)
```

```python
plt.subplots(figsize=(12,8))
plt.scatter(y_test, predicted)
plt.show()
```
